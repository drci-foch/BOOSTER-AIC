```python
import os
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
import imageio
```

```python
test_dir = "D:\\data_ETIS_781\\Training\\Test_dataset\\SWI_Test"
image_number = 55

image_filename = os.listdir(test_dir)[image_number-1]
image_data = nib.load(os.path.join(test_dir, image_filename)).get_fdata()

label_dir = "D:\\data_ETIS_781\\Training\\Test_labels\\MASK_Test"

label_filename = os.listdir(label_dir)[image_number-1]
label_data = nib.load(os.path.join(label_dir, label_filename)).get_fdata()

prediction_dir = "D:\\data_ETIS_781\\Training\\Predictions\\Test_Predictions\\SWI_Predictions\\Predictions_patients500_gamma5_sigmoid05"

prediction_filename = os.listdir(prediction_dir)[image_number-1]
prediction_data = nib.load(os.path.join(prediction_dir, prediction_filename)).get_fdata()
```

```python
voxel_max = np.max(image_data)
image_number = 2


for slice in range(image_data.shape[2]):
    frame_name = image_filename.split(".")[0]
    image_2D = (image_data[:,:,slice] / voxel_max).T
    plt.imshow(image_2D, cmap="gray", origin="lower")
    plt.axis("off")
    plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
    plt.savefig(f"D:\\data_ETIS_781\\Animations\\Animations_patients500_gamma5_sigmoid05\\Image_{image_number}\\Original\\{frame_name}_Original_Animation_Frame{slice}.png", transparent=False)
    plt.close()

for slice in range(image_data.shape[2]):
    frame_name = image_filename.split(".")[0]
    image_2D = (image_data[:,:,slice] / voxel_max).T
    image_rgb = np.stack([image_2D] * 3, axis=-1)
    overlay_label = image_rgb.copy()
    overlay_label[label_data[:,:,slice].T == 1] = [255/255, 120/255, 0]
    plt.imshow(overlay_label, cmap="gray", origin="lower")
    plt.axis("off")
    plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
    plt.savefig(f"D:\\data_ETIS_781\\Animations\\Animations_patients500_gamma5_sigmoid05\\Image_{image_number}\\Ground_Truth\\{frame_name}_Label_Animation_Frame{slice}.png", transparent=False)
    plt.close()

for slice in range(image_data.shape[2]):
    frame_name = image_filename.split(".")[0]
    image_2D = (image_data[:,:,slice] / voxel_max).T
    image_rgb = np.stack([image_2D] * 3, axis=-1)
    overlay_prediction = image_rgb.copy()
    overlay_prediction[prediction_data[:,:,slice].T == 1] = [255/255, 120/255, 0]
    plt.imshow(overlay_prediction, cmap="gray", origin="lower")
    plt.axis("off")
    plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
    plt.savefig(f"D:\\data_ETIS_781\\Animations\\Animations_patients500_gamma5_sigmoid05\\Image_{image_number}\\Prediction\\{frame_name}_Prediction_Animation_Frame{slice}.png", transparent=False)
    plt.close()
```

```python
image_number = 55
frames = []
frame_name = image_filename.split(".")[0]

for frame in range(image_data.shape[2]):
    image = imageio.v3.imread(f"D:\\data_ETIS_781\\Animations\\Animations_patients500_gamma5_sigmoid05\\Image_{image_number}\\Original\\{frame_name}_Original_Animation_Frame{frame}.png")
    frames.append(image)

imageio.mimsave(f"D:\\data_ETIS_781\\Animations\\Animations_patients500_gamma5_sigmoid05\\Image_{image_number}\\Original\\{frame_name}_Original_Animation.gif", frames, fps=8, loop=0)

frames = []
for frame in range(image_data.shape[2]):
    image = imageio.v3.imread(f"D:\\data_ETIS_781\\Animations\\Animations_patients500_gamma5_sigmoid05\\Image_{image_number}\\Ground_Truth\\{frame_name}_Label_Animation_Frame{frame}.png")
    frames.append(image)

imageio.mimsave(f"D:\\data_ETIS_781\\Animations\\Animations_patients500_gamma5_sigmoid05\\Image_{image_number}\\Ground_Truth\\{frame_name}_Label_Animation.gif", frames, fps=8, loop=0)

frames = []
for frame in range(image_data.shape[2]):
    image = imageio.v3.imread(f"D:\\data_ETIS_781\\Animations\\Animations_patients500_gamma5_sigmoid05\\Image_{image_number}\\Prediction\\{frame_name}_Prediction_Animation_Frame{frame}.png")
    frames.append(image)

imageio.mimsave(f"D:\\data_ETIS_781\\Animations\\Animations_patients500_gamma5_sigmoid05\\Image_{image_number}\\Prediction\\{frame_name}_Prediction_Animation.gif", frames, fps=8, loop=0)
```
